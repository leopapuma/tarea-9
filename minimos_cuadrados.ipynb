```python
#librerias que se usaran
import sympy as sp
import math

# Defenimos la aproximacion resultante
x = sp.symbols('x')
u_aprox = -sp.Rational(24759,173578)*x**3 - sp.Rational(1335,86789)*x**2 + 2*x


print("u_N(x) simbólico =", sp.simplify(u_aprox))

# escribimos los Coeficientes numéricos
a3 = sp.Rational(-24759,173578)
a2 = sp.Rational(-1335,86789)
print("a0 = 0")
print("a1 = 2")
print("a2 =", a2, "≈", float(a2))
print("a3 =", a3, "≈", float(a3))

# Error simbólico que obtuvimos
E_expr = -sp.Rational(242515,86789)*sp.sin(1) - sp.sin(2)/4 + sp.Rational(321213232883,451939831260) + sp.Rational(300043,86789)*sp.cos(1)
print("Error simbólico E =", sp.simplify(E_expr))

# Error numérico
E_num = float(sp.N(E_expr, 20))
print("Error numérico E ≈", E_num)
```
